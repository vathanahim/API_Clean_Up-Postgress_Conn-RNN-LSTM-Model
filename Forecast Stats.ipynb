```python
import tsfeatures as tf
```

```python
features = df_predict.columns[1:48]
features
```

```python
def get_value(feature):
    values = df_predict[feature].values
    return values


def get_stats(feature_values):
    entropy = tf.entropy(feature_values)
    tsf_stl_features = tf.stl_features(feature_values)
    return entropy, tsf_stl_features
```

```python
stats_dic_entropy = {}
stats_dic_stl = {}
stats_dic = {}
for i in features:
    values = get_value(i)
    entropy, tsf_stl_features = get_stats(values)
    stats_dic_entropy[str(i)+" entropy"] = entropy
    stats_dic_stl[str(i)+" stl_features"] = tsf_stl_features
```

```python
df_entropy = pd.DataFrame.from_dict(stats_dic_entropy,orient='index')
df_stl = pd.DataFrame.from_dict(stats_dic_stl,orient='index')
```

```python
import statsmodels.api as sm
from plotly.subplots import make_subplots
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.express as px
```

```python
def decom_plot(feature):
    
    df_index = df_predict[['dates',feature]]
    df_index = df_index.set_index(df_index['dates'])
    df_index = df_index[[feature]]
    decomposition = sm.tsa.seasonal_decompose(df_index, model='additive', period=7)
    fig = make_subplots(rows=3, cols=1)
    fig1 = px.line(decomposition.trend, title="hello")
    fig2 = px.line(decomposition.seasonal, title=feature)
    fig3 = px.line(decomposition.observed, title=feature)
    fig.add_trace(fig1['data'][0], row=1, col=1)
    fig.add_trace(fig2['data'][0], row=2, col=1)
    fig.add_trace(fig3['data'][0], row=3, col=1)
    fig.update_layout(height=600, width=800, title_text=i)
    return fig.show()
```
