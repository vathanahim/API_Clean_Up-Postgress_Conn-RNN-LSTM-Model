### Cleaning

```python
df = pd.read_excel(r'\Users\vhim\Documents\apple.xlsm', sheet_name='1B OB FG Feeder')
```

```python
df.dropna(subset=['Channel_Cd (group)'],inplace=True)
col_list = ['Channel_Cd (group)','Capacity LOB ID', 'Measure Values', 'Rptg_Dt']
select_df=df[col_list]
select_df.columns = ['channel_cd','entity', 'units', 'period']
select_df['entity'] = select_df['entity'].str.upper()
select_df['efa_id'] = 3
select_df['edit_who'] = 'INGESTDCFCSTJOB'
select_df['add_who'] = 'INGESTDCFCSTJOB'
```

```python
select_df['entity'] = select_df['entity'].replace({'ACCESSORIES':'ACCESSORY', 'DESKTOP':'MAC'})
```

### Joining

```python
from sqlalchemy import create_engine
```

```python
conn = create_engine('postgresql://datasci:pswd@serv:5433/datasci')
connection = conn.raw_connection()
cursor = connection.cursor()
```

```python
sql = """
select  lh.name lob, lh.lhid 
from dev.level_hierarchy lh
join dev.level l
on l.lid = lh.lid
where l.efa_id = 3
and level = 1

"""
```

```python
df_sql = pd.read_sql(sql, conn)
```

```python
df_sql
```

```python
#join will leave OUT IUP and MOBILE
df_join = select_df.merge(df_sql, how='inner', left_on='entity', right_on='lob')
```

```python
df_join.dtypes
```

```python
pd.unique(df_join[['entity','lhid']].values.ravel('K'))
```

```python
dfinal = df_join[['efa_id', 'lhid', 'entity', 'units', 'period', 'add_who', 'edit_who']]
dfinal
```

```python
df_final = dfinal.groupby(['efa_id', 'lhid', 'entity','period', 'add_who', 'edit_who'], as_index=False)['units'].sum()
```

```python
df_final = df_final[['efa_id', 'lhid', 'entity', 'units', 'period', 'add_who', 'edit_who']] 
```

```python
df_final.head()
```

```python
df_final_list = df_final.values.tolist()
```

```python
df_final_list
```

### Insert

```python
connection.rollback()
try:
    cursor.executemany("""
    insert into dev.external_forecast (efa_id, lhid, entity, units, period, add_who, edit_who) 
    values (%s, %s, %s, %s, %s, %s, %s)
    """, df_final_list)
    connection.commit()
    print("Insert Success")
except Exception as e:
    print(str(e))
    print("Insert Fail")
```
